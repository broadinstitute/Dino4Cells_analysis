```python
import torch
import numpy as np
import pandas as pd
from tqdm import tqdm
from pathlib import Path
from skimage import io
from multiprocessing import Pool
```

```python
cell_files = [str(f) for f in Path('/scr/mdoron/allen/crop_seg/').glob('*.tif')]
```

```python
def preprocess_image(f):
    io.imsave(f.replace('crop_seg','crop_seg_maxed'), io.imread(f).max(axis=0))
```

```python
pool = Pool(40)
list(tqdm(pool.imap(preprocess_image, cell_files)))
```
